## **Mini Project – Marks Adding**

*   **Input file** contain Roll numbers and marks (raw data).
*   Create **output file** that add all the marks.



```python
import pandas as pd
```

```python
#input file
input_file = "/content/Mini Project - Marks Adding (1).csv"
ip = pd.read_csv(input_file)

#Grouping marks
grp_ip = ip.groupby('Roll Num', as_index= False)['Marks'].sum()

#writing in output_file
op="output.csv"
grp_ip.to_csv(op, index = False)

print(f"Data written to the '{op}' successfully.")
```
